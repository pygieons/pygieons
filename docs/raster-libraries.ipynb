# Raster libraries

## Libraries

```python
from pygieons import Ecosystem

# Initialize
e = Ecosystem(plot_type="raster", log=False)

# Prepare the table and plot it
e.prepare_table().show()
```

## Linkages

```python
# Initialize
e = Ecosystem(plot_type="raster+generic", log=False)

# Prepare network and plot
net = e.prepare_net()
net.show()
```
